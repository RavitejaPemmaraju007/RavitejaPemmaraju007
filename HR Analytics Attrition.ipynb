```python
cols = ['NumCompaniesWorked', 'PerformanceRating','StockOptionLevel','YearsAtCompany','TrainingTimesLastYear','YearsSinceLastPromotion'] # one or more

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

## Data Preparation

### Conversion of categorical variable to Numerical variables

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

def Encoder(df):
    obj_cols = list(df.select_dtypes(include = 'object'))
    le = LabelEncoder()
    for cols in obj_cols:
        df[cols] = le.fit_transform(df[cols])
    return df
df = Encoder(df)
```

```python
df.dtypes
```

```python
df.head(10)
```

### Rescaling the features

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
vars = ['Age','BusinessTravel','DailyRate','Department','DistanceFromHome','Education','EducationField','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobRole','JobSatisfaction','MaritalStatus','MonthlyIncome','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']
df[vars] = scaler.fit_transform(df[vars])
```

```python
df.shape
```

```python
df.describe()
```

### Splitting the data into train and test splits

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Attrition'], 1)
X.head()
```

```python
# Putting the target variable in y
y = df['Attrition']
y.head()
```

```python
# Split the dataset into 90% and 10% for train and test respectively
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, test_size=0.1, random_state=10)
```

## Data Modeling

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
# Import RFE
from sklearn.feature_selection import RFE
```

```python
# Running RFE with 20 variables as output
rfe = RFE(lr, 20)
rfe = rfe.fit(X_train, y_train)
```

```python
# Features that have been selected by RFE
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# Put all the columns selected by RFE in the variable 'col'
col = X_train.columns[rfe.support_]
```

```python
# Selecting columns selected by RFE
X_train = X_train[col]
```

```python
lr.fit(X_train,y_train)
```

```python
col = X_train.columns
X_test = X_test[col]
```

```python
y_pred = lr.predict(X_test)
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,classification_report,roc_curve
```

```python
print('Accuracy score of the model is : {0:.2f}%'.format(100*accuracy_score(y_pred, y_test)))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
y_pred_prob = lr.predict_proba(X_test)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(y_test,y_pred_prob,drop_intermediate = False)
```

```python
plt.plot(fpr,tpr,label = 'Logistic Regression')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression Curve')
```
